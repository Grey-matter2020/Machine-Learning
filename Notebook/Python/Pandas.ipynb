```python
#fillna()
df = pd.DataFrame({'A' :[0, 3, None, 10, 3, None],'B' : [None, None, 7.13, 13.82, 7, 7],'C' : [None, "Pandas", None, "Pandas", "Python", "JavaScript"]}) 
print(df)
```

```python
#concat()
df1 = pd.DataFrame([['a', 1], ['b', 2]],columns=['letter', 'number'])
print (df1)

df2 = pd.DataFrame([['c', 3], ['d', 4]],columns=['letter', 'number'])
print (df2)
```

```python
pd.concat([df1, df2])                          # deault is axis =0
                                               # Notice the index
```

```python
pd.concat([df1, df2],ignore_index= True)                          
```

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]],columns=['letter', 'number'])
print (df1)

df4 = pd.DataFrame([['bird', 'polly'], ['monkey', 'george']],columns=['animal', 'name'])
print(df4)
```

```python
#groupby()
df = pd.DataFrame({'Animal': ['Falcon', 'Falcon','Parrot', 'Parrot'],
                   'Max Speed': [380., 370., 24., 26.]})
df
```

```python
df.groupby(['Animal']).mean()
```

```python
#merge()                 # same as joins in sql
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
print(df1
      )
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
print(df2)
```

```python
df1.merge(df2, how ='left', left_on='lkey', right_on='rkey')
```
